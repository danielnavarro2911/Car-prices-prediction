## Now, let's pay our attenntion on categorical feature

We have 147 unique names on CarName

```python
df.select_dtypes(include=['object']).apply(lambda x: x.nunique())
```

```python
df[['CarName']]
```

As we can see, CarName follows the structure: "Brand Model". Let's create a new column called "Brand"

```python
df['Brand']=df['CarName'].map(lambda x:x.split()[0].lower())
df.drop('CarName',axis=1,inplace=True)
```

Now, we can agroup this brands and see the price relation, using the mean value

```python
df.groupby('Brand').mean()[['price']].sort_values('price').plot.bar(figsize=(12,8))
```

Here, we can see there are some typo error that, shoud be manually fixed

```python
def brand(x):
    if x=='maxda':
        return 'mazda'
    if x=='porcshce':
        return 'porsche'
    if x=='toyouta':
        return 'toyota'
    if x=='vokswagen':
        return 'volkswagen'
    else:
        return x
        
```

```python
df['Brand']=df['Brand'].map(brand)
```

```python
df.groupby('Brand').mean()[['price']].sort_values('price').plot.bar(figsize=(12,8))
```
